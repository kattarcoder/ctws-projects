```r
content <- read.csv("train.csv")
```

```r
content
```

```r
data = as.vector(content)
```

```r
table(data$Gender)
```

```r
male <- sum(data$Gender == "Male", na.rm=TRUE)
female <- sum(data$Gender == "Female")
pie(c(male, female), c("Male", "Female"))
```

```r
table(data$Married)
```

```r
married <- sum(data$Married == "Yes", na.rm=TRUE)
not_married <- sum(data$Married == "No", na.rm=TRUE)
pie(c(married, not_married), c("married", "not_married"))
```

```r
table(data$Dependents)
```

```r
d0 <- sum(data$Dependents == "0", na.rm=TRUE)
d1 <- sum(data$Dependents == "1", na.rm=TRUE)
d2 <- sum(data$Dependents == "2", na.rm=TRUE)
d3 <- sum(data$Dependents == "3+", na.rm=TRUE)
pie(c(d0, d1, d2, d3), c("0", "1", "2", "3+"))
```
